### ë¬¸ì œ: ì „ ì§ì›ì˜ ì›”ê¸‰ ì´ì•¡ êµ¬í•˜ê¸°
  * ì§ê¸‰ì´ 2ê¸‰ ì´ìƒì´ë©´ ì›”ê¸‰ì´ ë³¸ë´‰+ìƒì—¬ê¸ˆ(ë³¸ë´‰ì˜ 30%)

  * í•©ê³„ ë³€ìˆ˜ total
  * ì§ê¸‰ ë³€ìˆ˜ rank ì…ë ¥ ë°›ê³

```python
empoloyee_num = int(input('ì§ì› ìˆ˜: '))
total = 0

for i in range(empoloyee_num): # 0,1,2 ê¹Œì§€ ì…ë ¥
  rank = int(input('ì§ê¸‰: '))
  basic_salary = int(input('ë³¸ë´‰: '))
  if rank >= 2:
    bonus = basic_salary * 0.3
  else:
    bonus = basic_salary * 0.2
  salary = basic_salary + bonus
  total += salary # total ë³€ìˆ˜ì— ì›”ê¸‰ í•© ì €ì¥..

print(f'ì „ì§ì› ì›”ê¸‰ ì´ì•¡: {total}')

```

### ë¬¸ì œ: ê²½ì§„ëŒ€íšŒ í‰ê°€ ì ìˆ˜ êµ¬í•˜ê¸°

* ì‹¬ì‚¬ìœ„ì› judge_num (5ëª…ì´ ì‹¬ì‚¬)
* score ë³€ìˆ˜ì— ê° ì ìˆ˜ë¥¼ ì…ë ¥ ë°›ê³ 
* ì ìˆ˜ ì´ í•©ê³„ sumì˜ 'í‰ê· 'ì„ ì¶œë ¥
* ìµœê³  ì ìˆ˜ëŠ” ë¹¼ê³  í•©ê³„ë¥¼ êµ¬í•´ì•¼ í•¨ max_score
* í˜„ì¬ ì ìˆ˜ê°€ ìµœê³  ì ìˆ˜ë³´ë‹¤ í¬ë©´,
  
  í˜„ì¬ ì ìˆ˜ë¥¼ max_scoreì— ì €ì¥í•˜ê³  sumì—ì„œ ë¹¼ì¤˜ì•¼ í•©ë‹ˆë‹¤.



```python
judge_num = int(input('ì‹¬ì‚¬ìœ„ì›ì˜ ìˆ˜: '))
sum = 0
max_score = 0

for i in range(judge_num): # 0,1,2,3,4
  score = int(input('ì ìˆ˜: ')) # 100, 80, 70, 90, 100
  sum += score # 100+80+70+90+100
if score > max_score:
  max_score = score # 100
sum -= max_score
print(f'ì‘í’ˆ í‰ê°€ ì ìˆ˜: {sum/judge_num-1}')
```

### ğŸ¯ ìˆ«ì ë§íˆê¸° ê²Œì„
1. **ì»´í“¨í„°ê°€ ìƒê°í•œ ìˆ˜ ë§íˆê¸°**:
    * `random.randint(1, 100)`ìœ¼ë¡œ ì •ë‹µì„ ìƒì„±í•©ë‹ˆë‹¤.
    * ì‚¬ìš©ìê°€ ì •ë‹µì„ ë§í ë•Œê¹Œì§€ `while` ë¬¸ìœ¼ë¡œ ì…ë ¥ì„ ë°˜ë³µí•©ë‹ˆë‹¤.
    * ì…ë ¥ê°’ì´ í¬ë©´ "ë„ˆë¬´ ë†’ì•„ìš”", ì‘ìœ¼ë©´ "ë„ˆë¬´ ë‚®ì•„ìš”"ë¥¼ ì¶œë ¥í•˜ì—¬ ë²”ìœ„ë¥¼ ì¢í˜€ê°‘ë‹ˆë‹¤.
2. **ì‚¬ìš©ìê°€ ìƒê°í•œ ìˆ˜ ë§íˆê¸°**:
    * ì‚¬ìš©ìê°€ ìˆ«ìë¥¼ í•˜ë‚˜ ì •í•˜ë©´ ì»´í“¨í„°ê°€ `low`ì™€ `high` ë³€ìˆ˜ë¥¼ ì¡°ì •í•˜ë©° ì •ë‹µì„ ì°¾ì•„ê°€ëŠ” ë¡œì§ì…ë‹ˆë‹¤.

```python
# 1. ì»´í“¨í„°ê°€ ìƒê°í•œ ìˆ«ì ë§ì¶”ê¸°
import random

print('1ë¶€í„° 100ê¹Œì§€ì˜ ìˆ«ì í•˜ë‚˜ë¥¼ ì •í–ˆìŠµë‹ˆë‹¤.')
print('ì´ ìˆ«ìë¥¼ ë§ì¶°ë³´ì„¸ìš”!')

answer = random.randint(1,100) # 1~100ì‚¬ì´ ìˆ«ì í•˜ë‚˜ ëœë¤ìœ¼ë¡œ ì €ì¥
time = 0 # ë„ì „ íšŸìˆ˜

# while ë¬¸ìœ¼ë¡œ ë°˜ë³µí•˜ëŠ”ë°(ì¦‰, ë„ì „ì´ ê³„ì†ë¨)
while True:
  user = int(input('ìˆ«ìë¥¼ ì…ë ¥í•˜ì„¸ìš”: '))
  time += 1 # ë„ì „ íšŸìˆ˜ 1 ì¦ê°€

# userê°€ ì…ë ¥í•œ ìˆ«ìê°€ ë§ìœ¼ë©´ ë°˜ë³µë¬¸ ì¢…ë£Œê°€ ë˜ê³ 
  if user == answer:
    break

# userê°€ ì…ë ¥í•œ ìˆ«ìê°€ í¬ë©´ í¬ë‹¤ê³ , ì‘ìœ¼ë©´ ì‘ë‹¤ê³  ì¶œë ¥
  elif user > answer:
    print(f'{user}ë³´ë‹¤ ì‘ìŠµë‹ˆë‹¤')

  else:
    print(f'{user}ë³´ë‹¤ í½ë‹ˆë‹¤')

# ìµœì¢… ë„ì „ ëª‡ íšŒë§Œì— ì •ë‹µì„ ë§ì·„ëŠ”ì§€ ì¶œë ¥
print(f'ì •ë‹µì…ë‹ˆë‹¤! {time}ë²ˆë§Œì— ë§ì·„ìŠµë‹ˆë‹¤')



```

```python
# 2. ì‚¬ìš©ìê°€ ìƒê°í•œ ìˆ˜ ë§íˆê¸°
print('1~100 ì‚¬ì´ ìˆ«ì í•˜ë‚˜ë¥¼ ì •í•˜ì„¸ìš”')
print('ì´ ìˆ«ìë¥¼ ë§í˜€ë³´ê² ìŠµë‹ˆë‹¤!')

low = 1
high = 100
times = 0

while True:
  num = (low+high)//2 # ì»´í“¨í„°ê°€ ì œì‹œí•œ ìˆ«ì
  answer = input(f'{num}ì¸ê°€ìš”? ok/up/down') # ì‚¬ìš©ìê°€ ë‹µë³€ì„ í•´ì¤Œ
  times += 1

  if answer == 'ok':
    break
  elif answer == 'up':
    low = num+1
  elif answer == 'down':
    high = num-1
  else:
    print(f'ì˜ëª» ì…ë ¥í–ˆìŠµë‹ˆë‹¤.')

print(f'{times}íšŒë§Œì— ë§í˜”ìŠµë‹ˆë‹¤')

```
