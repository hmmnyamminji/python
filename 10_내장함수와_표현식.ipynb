### **all() 함수** : 인자로 받은 값의 요소들 중 하나라도 False면 결과가 False 이며, 모든 요소가 True 일 경우 True이다.  and
### **any() 함수** : 인자로 받은 값의 요소들 중 하나라도 True 일 경우 결과가 True이며, 모두 False이면 결과가 False 이다. or

````python
num = 30
check_num = [num%2==0, num%3==0, num%5==0]
if all(check_num):
  print('all : and')
````

````python
num = 8
check_num = [num%2==0, num%3==0, num%5==0]
if any(check_num):
  print('any : or')
````

````python
print(all([True, True, False, True]))
print(all([True, True, True, True]))
print(any([False, False, False]))
print(any([True, False]))
````

### if문 확장
* **all(), any()** 를 사용한 **if문**
* **short circuit evaluation** : 먼저 등장한 논리식이 뒤의 논리식과 관계없이 전체 연산 결과를 결정할 경우 뒤의 논리식을 실행하지 않는다


```
if True or 조건 : 결과가 이미 True임으로 뒤의 논리식은 계산하지 않는다(=볼 것도 없다)
  실행코드
if False and 조건 : 결과가 이미 False임으로 뒤의 논리식을 계산하지 않는다
  실행코드
```



````python
if True or 3%2 == 0:
  print('True or')

if False and 3%2 == 0:
  print('False and')
````

````python
nums = [1, -5, 3]
if nums [0]>0:
  print('positive')
````

````python
nums = []
if nums[0]>0:
  print('positive') # 빈 리스트라서 에러
````

````python
nums = []
if len(nums)>0 and nums[0]>0: # False and 임으로 뒤에 논리식이 무시된다.
  print('positive')
````

````python
print(any([10//i > 0 for i in [3,2,1,0]]))
# 반복 대상 전체에 대해 계산하는데 리스트 마지막 숫자가 0이라서 0으로 나눌 수 없어서 에러남
````

````python
print(any(10//i>0 for i in [3,2,1,0]))
# 첫번째 반복 i=3에서 10//i>0이 True다.
# 이후 결과에 상관없이 any() 함수의 결과는 True가 되어 뒤의 검사를 수행하지 않는다.
#any(True)
````

* 시간 복잡도 : 입력의 크기에 따라 알고리즘이 실행되는 데에 필요한 시간의 양을 나타내는 개념. 빅오 표기법(Big O Notation)으로 표시,


```
O(1) 입력의 크기 n에 대해 관계없이 동일한 시간이 걸리는 복잡도 표현
O(n) 입력의 크기 n에 대해 선형적인 시간이 걸리는 복잡도 표현
O(nlogn) 입력의 크기 n에 대해 n*logn에 비례하는 시간이 걸리는 복잡도 표현 (log밑2)

O(n**2) 입력의 크기 n의 제곱에 비례하는 시간이 걸리는 복잡도 표현

O(n**3) 입력의 크기 n의 세제곱에 비례하는 시간이 걸리는 복잡도 표현
O(a**n) 입력의 크기 n에 지수적인 관계로 시간이 걸리는 복잡도 표현 (a>1)
```
