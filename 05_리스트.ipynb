```python
menus = ['ìì¥ë©´','ì§¬ë½•','ë³¶ìŒë°¥'] # ì¸ë±ìŠ¤ 0 1 2ë²ˆ ë°©
print(menus[1])

menus[2] = "íƒ•ìˆ˜ìœ¡" # ì¸ë±ìŠ¤ 2ë²ˆë°©ì— íƒ•ìˆ˜ìœ¡ì„ ì…ë ¥í•´ì„œ ë³¶ìŒë°¥ -> íƒ•ìˆ˜ìœ¡ ë°”ë€œ
print(menus[2])

print(menus)
print(menus[-1])
print(menus[-2])
print(menus[-3])

for i in range(len(menus)): #0,1,2
  print(menus[i])

for item in menus:
  print(item)
```
