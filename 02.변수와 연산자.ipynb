# 📘 파이썬 기초: 02. 변수와 연산자

## 1. 변수 (Variable)

* **개념**: 프로그램 실행 중 값을 저장하는 메모리 공간.
* **특징**: 변수에는 하나의 값만 저장되며, 새로운 값을 넣으면 기존 값은 삭제됨.
* **대입 연산자 (`=`)**: 오른쪽의 값을 왼쪽 변수에 저장함.
* 예: `age = 20` (age라는 변수에 20을 대입)

## 2. 자료형 (Data Type)

데이터의 종류에 따라 적절한 자료형을 사용해야 합니다.

* **정수형 (int)**: 소수점이 없는 숫자 (예: 10, -5)
* **실수형 (float)**: 소수점이 있는 숫자 (예: 3.14, 0.5)
* **문자열 (str)**: 따옴표(`'`, `"`)로 감싸진 텍스트 (예: "Hello", 'Python')
* **불형 (bool)**: 참(`True`) 또는 거짓(`False`)을 나타냄

## 3. 입력과 출력

* **출력 (`print`)**: 변수나 값을 화면에 표시.
* `print('age')`는 문자열 출력, `print(age)`는 변수 값 출력.

* **입력 (`input`)**: 사용자로부터 데이터를 입력받음.
* **주의**: `input()`으로 받은 데이터는 기본적으로 **문자열**로 처리되므로, 계산이 필요하면 `int()`나 `float()`로 변환해야 함.

## 4. 연산자 (Operators)

산술 연산자를 통해 데이터를 계산합니다.

| 연산자 | 기능 | 예시 (a=10, b=3) | 결과 |
| --- | --- | --- | --- |
| `+` | 더하기 | `a + b` | 13 |
| `-` | 빼기 | `a - b` | 7 |
| `*` | 곱하기 | `a * b` | 30 |
| `/` | 나누기 (실수) | `a / b` | 3.333... |
| `//` | **몫** | `a // b` | 3 |
| `%` | **나머지** | `a % b` | 1 |
| `**` | 거듭제곱 | `a ** b` | 1000 |

## 5. [실습 예제]
### ⏱ 시간 단위 변환기
1. **데이터 입력**: 사용자로부터 초(second) 단위의 큰 숫자를 입력받아 정수로 변환합니다.
2. **시간/분 계산**:
    * 전체 초를 3,600으로 나눈 **몫**을 '시간'으로 저장합니다.
    * 남은 초를 다시 60으로 나눈 **몫**은 '분', **나머지**는 최종적인 '초'가 됩니다.
3. **결과 출력**: 계산된 시간, 분, 초를 조합하여 화면에 보여줍니다.



### 💰 주문 금액 계산기
1. **정보 입력**: 상품 가격과 주문 수량을 각각 입력받아 숫자로 변환합니다.
2. **금액 산출**: (상품 가격 × 수량) 로직을 통해 총 주문 금액을 계산합니다.
3. **포맷팅**: "총 주문 금액은 OOO원입니다"와 같은 형태로 가독성 있게 출력합니다.
