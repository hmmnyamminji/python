### 분위수 (Quantile / Percentile)

***데이터를 정렬했을 때 일정 비율의 위치에 해당하는 값***

```python
# 분위수 구하기
print("=== 분위수 (25%, 50%, 75%) ===")
print(df.quantile([0.25, 0.5, 0.75], numeric_only=True))
```

```python
# 특정 칼럼의 분위수
col = 'sepal_length'
Q1 = df[col].quantile(0.25)
Q2 = df[col].quantile(0.50)
Q3 = df[col].quantile(0.75)
IQR = Q3 - Q1
print(f"₩n꽃받침 길이 분위수: ")
print(f" Q1 (25%): {Q1}")
print(f" Q1 (50%): {Q2}")
print(f" Q1 (75%): {Q3}")
print(f" IQR  : {IQR}")
```

```python
# 이상치 경계값 (정상 범위를 벗어났다고 판단하는 기준선)
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
print(f"₩n이상치 경계값:")
print(f" 하한: {lower:.2f}")
print(f" 상한: {upper:.2f}")
```

```python
# 이상치 확인
outliers = df[(df[col] < lower) | (df[col] > upper)]
print(f"₩n 이상치 데이터 수: {len(outliers)}개")
# df[True, False, True] DataFrame에 불리언 시리즈를 전달하면 True로 표시된 행들만 필터
# True로 표시된 행들만 필터링하여 새로운 DataFrame인 outliers를 반환
```

### 분산(Variance)
***각 데이터 값이 평균으로부터 얼마나 퍼져 있는지를 나타내는 지표***

```python
# 단일 칼럼 분산
# print("꽃받침 길이 분산:", df['sepal_length'].var())
print(f"꽃받침 길이 분산: {df['sepal_length'].var():.2f}")
```

```python
# 전체 칼럼 분산
print("₩n=== 전체 칼럼 분산 ===")
print(df.var(numeric_only=True).round(2))
```

```python
# 종류별 분산 비교 (분산이 크면 데이터가 넓게 퍼져 있음)
print("₩n종류별 꽃잎 길이 분산:")
print(df.groupby('species')['petal_length'].var().round(2))
```

### 표준편차 (Standard Deviation) ###

**분산의 제곱근**입니다. 분산과 달리 **원래 데이터와 같은 단위**여서 직관적으로 해석할 수 있습니다.

```python
# 단일 칼럼 표준편차
#print("꽃받침 길이 표준편차:", df['sepal_length'].std)
print(f"꽃받침 길이 표준편차: {df['sepal_length'].std():.2f}")
```

```python
# 전체 칼럼 표준편차
print("₩n=== 전체 칼럼 표준편차 ===")
print(df.std(numeric_only=True).round(2))
```

```python
# 분산 vs 표준편차 비교 정리
print("₩n=== 분산 vs 표준편차 비교 ===")
summary = pd.DataFrame({
    '분산': df.var(numeric_only=True).round(4),
    '표준편차': df.std(numeric_only=True).round(4)
})
print(summary)
# 표준편차(Standard Deviation): 분산의 제곱근으로,
# 데이터가 평균으로부터 평균적으로 얼마나 떨어져 있는지를 나타냅니다
# 데이터의 원래 단위를 따르기 때문에 분산보다 직관적으로 이해하기 쉽습니다.

#petal_length(꽃잎 길이): 평균으로부터 가장 넓게 퍼져 있으며, 데이터의 포인트들이 서로 크게 다르다는 것을 의미
#sepal_width(꽃받침 너비): 데이터가 평균 주변에 가장 밀집되어 있으며, 데이터의 포인트들이 서로 유사하다는 것을 의미
```

```python
# 평균 +- 표준편차 범위 (데이터의 약 68%가 이 범위에 속함)
col = 'sepal_length'
mean = df[col].mean()
std = df[col].std()
print(f"₩n꽃받침 길이: 평균 +- 표준편차")
print(f" 평균: {mean:.2f}")
print(f" 표준편차: {std:.2f}")
print(f" 범위: {mean - std:.2f} ~ {mean + std:.2f}")
count_in_range = df[(df[col] >= mean - std) & (df[col] <= mean + std)].shape[0]
#True로 표시된 행들만 필터링하여 새로운 DataFrame인 count_in_range을 반환
# .shape[0]: 필터링된 DataFrame의 행의 개수를 반환
print(f" 이 범위에 속하는 데이터: {count_in_range}개 ({count_in_range/len(df)*1}%)")
```

### 데이터프레임에서 기술통계 한번에 구하기

```python
# 기본 기술통계 (숫자형 칼럼 자동 선택)
print("=== 기술통계 요약 (describe) ===")
print(df.describe().round(2))
```

```python
# 범주형 포함 전체 기술통계
print("₩n=== 범주형 포함 기술통계 ===")
print(df.describe(include='all').round(2))
```

```python
# 종류별 기술통계 (groupby + describe)
print("₩n=== 종류별 기술통계 ===")
print(df.groupby('species').describe())
```

```python
# 내가 원하는 통계만 선택해서 보기
print("₩n=== 커스텀 기술통계 ===")
custom_stats = df.groupby('species')['petal_length'].agg([
    ('평균', 'mean'),
    ('중앙값', 'median'),
    ('표준편차', 'std'),
    ('최솟값', 'min'),
    ('최댓값', 'max'),
]).round(3)
print(custom_stats)
#.agg([...]): 하나 이상의 집계(aggregation) 함수를 적용하는 메서드
# ('평균','mean'): 'mean' 함수를 적용하여 평균을 계산하고, 결과 칼럼의 이름을 '평균'으로 지정
```
