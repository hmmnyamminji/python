### 반복문 - loop

```python
words = ["apple","banana","grape"]
for item in words:
  print(item)
```

```python
nums = [5,4,3]
for i in nums:
  print(i*3)
```

```python
for i in range(3):
  print(i)
```

```python
for i in range(3):
  print(i*2)
```

```python
for i in range(1,3):
  print(1*3)
```

```python
nums = [10,20,30]
for i in range(len(nums)):
  print(nums[i])
```

```python
for i in range(5):
  if i % 2 == 0:
    print(i)
```

```python
words = ["ape", "monkey","gorilla"]
for w in words:
  if len(w) > 5:
    print(w)
```

```python
for i in range(3):
  for j in "ABC":
    print(i, j)
```

```python
nums = [10,20,30]
for i in range(len(nums)):
  for j in range(len(nums)):
    print(nums[i] * nums[j])
```

```python
i = 1
while i < 5:
  i += 2
  print(i)
```

```python
nums = []
i=1
while sum(nums) < 20:
  nums.append(i)
  i += 1
print(nums)
```

```python
for i in range(10):
  if i == 3:
    break
  print(i)
```

```python
nums = [1]
while True:
  i = sum(nums)
  nums += [i]
  if sum(nums) > 10:
    break
print(nums)
```

```python
for i in range(3):
  for j in range(3):
    print(j*2)
break
```

```python
for i in range(3):
  for j in range(3):
    print(j*2)
    break
```

```python
for i in range(3):
  if i == 10:
    break
  print(i)
else:
  print("no break")
```
