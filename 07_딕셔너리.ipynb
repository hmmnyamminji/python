* **ìŒìœ¼ë¡œ ê°€ì ¸ì˜¤ê¸° (items)**: í‚¤ì™€ ê°’ì˜ ìŒì„ íŠœí”Œ í˜•íƒœë¡œ ë¬¶ì–´ì„œ ë°˜í™˜í•˜ë©°, ë°˜ë³µë¬¸ê³¼ í•¨ê»˜ ìì£¼ ì‚¬ìš©ë©ë‹ˆë‹¤.

```python
# êµ­ê°€ ì´ë¦„ì„ ì…ë ¥ ë°›ê³  êµ­ê°€ì˜ ìˆ˜ë„ë¥¼ ì•Œë ¤ì£¼ëŠ” ì‹
country_capital = {'í•œêµ­':'ì„œìš¸', 'ì¼ë³¸':'ë„ì¿„', 'ì¤‘êµ­':'ë² ì´ì§•'}
country = input('êµ­ê°€ì˜ ì´ë¦„ì„ ì…ë ¥í•˜ì„¸ìš”: ')
if country in country_capital:
  print(country_capital[country])
else:
  print('f{country}ëŠ” ë“±ë¡ë˜ì§€ ì•Šì•˜ìŠµë‹ˆë‹¤')
```

```python
# ìŒë£Œ ì¢…ë¥˜ì™€ ì£¼ë¬¸ ìˆ˜ë¥¼ ì…ë ¥ ë°›ì•„ ì£¼ë¬¸ ê¸ˆì•¡ ê³„ì‚°í•˜ê¸°
prices = {'ì•„ë©”ë¦¬ì¹´ë…¸':2000, 'ë¼ë–¼':3000, 'ë…¹ì°¨': 2500}
sum = 0
while True:
  s = input('ìŒë£Œ ì¢…ë¥˜ ë˜ëŠ” ì¢…ë£Œ: ')
  if s == 'ì¢…ë£Œ':
    break
  elif s in prices:
    n = int(input(f'{s} ì£¼ë¬¸ ìˆ˜: '))
    sum += prices[s] * n
  else:
    print(f'{s}ëŠ” ì—†ëŠ” ìŒë£Œì…ë‹ˆë‹¤')

print(f'ì´ ê¸ˆì•¡: {sum}ì› ì…ë‹ˆë‹¤')
```

### ğŸ“Š í”„ë¡œê·¸ë˜ë° ì–¸ì–´ ì„ í˜¸ë„ ì¡°ì‚¬
1. **ë°ì´í„° ìˆ˜ì§‘**: ì‚¬ìš©ìë¡œë¶€í„° ì‚¬ìš© ê°€ëŠ¥í•œ í”„ë¡œê·¸ë˜ë° ì–¸ì–´ë¥¼ ì—¬ëŸ¬ ë²ˆ ì…ë ¥ë°›ìŠµë‹ˆë‹¤.
2. **ì¹´ìš´íŒ… ë¡œì§**:
    * ì…ë ¥ë°›ì€ ì–¸ì–´ê°€ ì´ë¯¸ ë”•ì…”ë„ˆë¦¬ì˜ í‚¤ë¡œ ì¡´ì¬í•˜ë©´, í•´ë‹¹ í‚¤ì˜ ê°’ì„ 1 ì¦ê°€ì‹œí‚µë‹ˆë‹¤.
    * ì¡´ì¬í•˜ì§€ ì•ŠëŠ” ìƒˆë¡œìš´ ì–¸ì–´ë¼ë©´, í•´ë‹¹ ì–¸ì–´ë¥¼ í‚¤ë¡œ ì¶”ê°€í•˜ê³  ê°’ì— 1ì„ ì €ì¥í•©ë‹ˆë‹¤.
3. **ê²°ê³¼ ì¶œë ¥**: ê° ì–¸ì–´ë³„ë¡œ ëª‡ ë²ˆì”© ì…ë ¥ë˜ì—ˆëŠ”ì§€ ì§‘ê³„ëœ ë°ì´í„°ë¥¼ ë³´ì—¬ì¤ë‹ˆë‹¤.

```python
langs = {'Python':0, 'C':0, 'Java':0, 'JavaScript':0}

while True:
  s = input('ì‚¬ìš©í•˜ëŠ” ì–¸ì–´ë¥¼ ì…ë ¥í•˜ì„¸ìš” ë˜ëŠ” ì¢…ë£Œ: ')
  if s == 'ì¢…ë£Œ':
    print('ì¢…ë£Œí•©ë‹ˆë‹¤')
    break
  elif s in langs:
    langs[s] += 1
  else:
    print('ì˜ëª» ì…ë ¥í–ˆìŠµë‹ˆë‹¤')

  for key, value in langs.items():
    print(key, value)
```

```python
# ì‚¬ìš© ê°€ëŠ¥í•œ í”„ë¡œê·¸ë˜ë° ì–¸ì–´ ì¡°ì‚¬
langs = {}

while True:
  s = input('ì‚¬ìš©í•˜ëŠ” ì–¸ì–´ë¥¼ ì…ë ¥í•˜ì„¸ìš”(ì‰¼í‘œ, ë¡œ êµ¬ë¶„í•´ì„œ ì…ë ¥) ë˜ëŠ” ì¢…ë£Œ: ')
  s = s.lower()
  if s == 'ì¢…ë£Œ':
    print('ì¢…ë£Œí•©ë‹ˆë‹¤')
    break
  #s = s.split(',')
  #print(s)

  for a in s.split(','):
    if a in langs:
      langs[a] += 1
    else:
      langs[a] = 1
  for key, value in langs.items():
    print(key, value)
```

### ğŸ¦ ì˜ì–´ ë‹¨ì–´ í€´ì¦ˆ (ë™ë¬¼ ì´ë¦„)
