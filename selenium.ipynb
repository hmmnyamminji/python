```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup

# 크롬 옵션
chrome_options = Options()
chrome_options.add_experimental_option("detach", True)

# 드라이버 실행
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)

print("드라이버 실행 완료")

# 올리브영 스킨케어 페이지
main_url = "https://www.oliveyoung.co.kr/store/display/getMCategoryList.do?dispCatNo=10000010001"
driver.get(main_url)

# 페이지 로딩 대기
wait = WebDriverWait(driver, 10)

wait.until(
    EC.presence_of_all_elements_located((By.CSS_SELECTOR, ".prd_info"))
)

time.sleep(2)

# HTML 파싱
soup = BeautifulSoup(driver.page_source, "html.parser")

# 상품 리스트
items = soup.select(".prd_info")

print("상품 개수:", len(items))

# 상품 링크 수집
sub_url = []

for item in items:

    a_tag = item.select_one("a")

    if a_tag:
        link = a_tag.get("href")

        if link:
            if link.startswith("http"):
                sub_url.append(link)
            else:
                sub_url.append("https://www.oliveyoung.co.kr" + link)

print("수집된 URL 수:", len(sub_url))

# 상세 페이지 크롤링
result = []

for i, url in enumerate(sub_url[:10], 1):

    print(f"[{i}/10] 크롤링 중")

    try:
        driver.get(url)

        wait.until(
            EC.presence_of_element_located((By.CSS_SELECTOR, ".prd_name"))
        )

        time.sleep(1)

        soup = BeautifulSoup(driver.page_source, "html.parser")

        name = soup.select_one(".prd_name")
        price = soup.select_one(".price_real")

        name = name.text.strip() if name else "상품명 없음"
        price = price.text.strip() if price else "가격 없음"

        result.append({
            "상품명": name,
            "가격": price,
            "url": url
        })

    except:
        print("페이지 오류 발생")

# 결과 출력
print("\n크롤링 결과")

for r in result:
    print(r)

driver.quit()
```

```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup

# 크롬 드라이버 실행
chrome_options = Options()
chrome_options.add_experimental_option("detach", True)

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)

# 올리브영 스킨케어 페이지 접속
driver.get("https://www.oliveyoung.co.kr/store/display/getMCategoryList.do?dispCatNo=10000010001")
time.sleep(3)

# 상품 목록 가져오기
soup = BeautifulSoup(driver.page_source, "html.parser")
items = soup.select(".prd_info")

# 상세페이지 URL 리스트 생성
sub_url = []

for item in items:
    a = item.select_one("a")

    if a and "href" in a.attrs:
        link = a["href"]

        if link.startswith("http"):
            sub_url.append(link)

        elif link.startswith("/"):
            sub_url.append("https://www.oliveyoung.co.kr" + link)
```
