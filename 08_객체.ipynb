## 2. í´ë˜ìŠ¤ ìƒì„± ë° ì‚¬ìš©

```python
class Account:
  def __init__(self, amount = 0):
    self.balance = amount #__balanceëŠ” private ì†ì„±

  def deposit(self, amount):
    self.balance += amount
    return self.balance

  def withdraw(self, amount):
    self.balance -= amount
    return self.balance

a1 = Account(50)
print(a1.__balance) # ì™¸ë¶€ì—ì„œ ì ‘ê·¼ ë¶ˆê°€
```

```python
# private ì†ì„± ê°’ì„ ì–»ìœ¼ë ¤ë©´ ì†ì„± ê°’ì„ ë°˜í™˜í•˜ëŠ” get_balance() ë©”ì„œë“œë¥¼ ì´ìš©í•´ì•¼ ë¨
class Account:
  def __init__(self, amount = 0):
    self.__balance = amount #__balanceëŠ” private ì†ì„±

  def deposit(self, amount):
    self.__balance += amount
    return self.__balance

  def withdraw(self, amount):
    self.__balance -= amount
    return self.__balance

  def get_balance(self):
    return self.__balance

a1 = Account(50)
print(a1.get_balance())
```

```python
# í´ë˜ìŠ¤ Circle ë§Œë“¤ê¸°
# ë°˜ì§€ë¦„ ì†ì„±__radius
# ë°˜ì§€ë¦„ ì†ì„±ì„ í™•ì¸í•˜ëŠ” ë©”ì„œë“œ get_radius
# ë°˜ì§€ë¦„ ì„¤ì •í•˜ëŠ” set_radius ë©”ì„œë“œ
# ì›ì˜ ë„“ì´ êµ¬í•˜ëŠ” get_area ë©”ì„œë“œ
# ì›ì˜ ë‘˜ë ˆë¥¼ êµ¬í•˜ëŠ” get_circleference ë©”ì„œë“œ

import math

class Circle:
  def __init__(self, radius=10):
    self.__radius = radius

  def get_radius(self):
    return self.__radius

  def set_radius(self, radius):
    self.__radius = radius

  def get_area(self):
    return math.pi*self.__radius**2

  def get_circumference(self):
    return 2*math.pi*self.__radius

c1 = Circle(7)
print(f'ë°˜ì§€ë¦„:{c1.get_area()}')
c1.set_radius(15)
print(f'ë°˜ì§€ë¦„:{c1.get_area()}')
print(f'ë„“ì´:{c1.get_area()}')
print(f'ë‘˜ë ˆ:{c1.get_circumference():.2f}')
```

## 3. ìƒì† (Inheritance)
* **ê°œë…**: ê¸°ì¡´ í´ë˜ìŠ¤(ë¶€ëª¨ í´ë˜ìŠ¤)ì˜ ì†ì„±ê³¼ ë©”ì„œë“œë¥¼ ê·¸ëŒ€ë¡œ ë¬¼ë ¤ë°›ì•„ ìƒˆë¡œìš´ í´ë˜ìŠ¤(ìì‹ í´ë˜ìŠ¤)ë¥¼ ë§Œë“œëŠ” ê¸°ëŠ¥ì…ë‹ˆë‹¤.
* **ì¥ì **: ì½”ë“œì˜ ì¬ì‚¬ìš©ì„±ì„ ë†’ì—¬ì£¼ë©°, ì¤‘ë³µëœ ì½”ë“œë¥¼ ì¤„ì—¬ ìœ ì§€ë³´ìˆ˜ê°€ ì‰¬ì›Œì§‘ë‹ˆë‹¤.
* **ë©”ì„œë“œ ì˜¤ë²„ë¼ì´ë”©(Overriding)**: ë¶€ëª¨ í´ë˜ìŠ¤ë¡œë¶€í„° ë¬¼ë ¤ë°›ì€ ë©”ì„œë“œë¥¼ ìì‹ í´ë˜ìŠ¤ì—ì„œ ìš©ë„ì— ë§ê²Œ ìƒˆë¡­ê²Œ ì¬ì •ì˜í•˜ì—¬ ì‚¬ìš©í•˜ëŠ” ê²ƒì„ ë§í•©ë‹ˆë‹¤.

```python
# ë¶€ëª¨ í´ë˜ìŠ¤
class Student:
  def __init__(self, name, computer):
    self.__name = name
    self.computer = computer

  def set_computer(self, computer):
    self.computer = computer

  def get_name(self):
    return self.__name
```

```python
# ìì‹ í´ë˜ìŠ¤
class Science(Student):
  def __init__(self, name, computer, science):
    super().__init__(name, computer)
    self.science = science

  def set_science(self, science):
    self.science = science

  def get_average(self):
    return (self.computer + self.science) / 2
```

```python
# ìì‹ í´ë˜ìŠ¤
class Social(Student):
  def __init__(self, name, computer, social):
    super().__init__(name, computer)
    self.social = social

  def set_social(self, social):
    self.social = social

  def get_average(self):
    return (self.computer + self.social) / 2
```

```python
st1 = Science('lee', 90, 80)
print(st1.get_name())
print(st1.get_average())
st1.set_computer(100)
print(st1.get_average())
```

```python
st2 = Social('kim', 98, 88)
print(st2.get_name())
print(st2.get_average())
st2.set_social(96)
print(st2.get_average())
```

```python
# ë¶€ëª¨ í´ë˜ìŠ¤
class Figure:
  def __init__(self, area, perimeter):
    self.area = area
    self.perimeter = perimeter

  def get_area(self):
    return self.area

  def get_perimeter(self):
    return self.perimeter
```

```python
# ìì‹ í´ë˜ìŠ¤
class Circle(Figure):
  def __init__(self, radius):
    area = 3.14*radius**2
    perimeter = 2*3.14*radius
    super().__init__(area, perimeter)
    self.radius = radius

  def get_radius(self, radius):
    self.radius = radius
    self.area = 3.14*radius**2
    self.perimeter = 2*3.14*radius

  def get_radius(self):
    return self.radius
```

```python
# ìì‹ í´ë˜ìŠ¤
class Rectangle(Figure):
  def __init__(self, width=10, height=10):
    area = width*height
    perimeter = 2*(width+height)
    super().__init__(area.perimeter)
    self.width = width
    self.height = height

  def set_width(self, width):
    self.width = width
    self.area = width*self.height
    self.perimeter = 2*(self.width+self.height)

  def set_height(self, height):
    self.height = height
    self.area = self.width*height
    self.perimeter = 2*(self.width+self.height)

  def get_width(self):
    return self.width

  def get_height(self):
    return self.height
```

```python
f1 = Circle(7)
f1.set_radius(70)
print(f1.get_area())
print(f1.get_area(), f1.get_perimeter())
```

```python
f2 = Rectangle(5,4)
print(f2.get_area(), f2.get_perimeter())
f2.set_width(10)
print(f2.get_area(), f2.get_perimeter())
f2.set_height(10)
print(f2.get_area(), f2.get_perimeter())
```

4. **ê°ì²´ ìƒì„±**: ì •ì˜ëœ í´ë˜ìŠ¤ë¥¼ ë°”íƒ•ìœ¼ë¡œ `car1`, `car2` ë“± ì„œë¡œ ë‹¤ë¥¸ ëª¨ë¸ëª…ê³¼ ì—°ë¹„ë¥¼ ê°€ì§„ ì‹¤ì œ ìë™ì°¨ ê°ì²´ë“¤ì„ ë§Œë“¤ì–´ ì‹¤í–‰í•©ë‹ˆë‹¤.

```python
class Car:
  def __init__(self, model, fuelefficiency):
    self.__model = model
    self.__fuelefficiency = fuelefficiency

  def get_model(self):
    return self.__model

  def get_fuelefficiency(self):
    return self.__fuelefficiency

  def set_fuelefficiency(self, fuelefficiency): # Renamed to set_fuelefficiency
    self.__fuelefficiency = fuelefficiency

  def get_fuelamount(self, distance):
    return distance/self.__fuelefficiency
```

```python
car1 = Car('GV70', 12)
print(f'ëª¨ë¸ëª…: {car1.get_model()}')
print(f'ì—°ë¹„: {car1.get_fuelefficiency()}')
distance = float(input('ì£¼í–‰ ê±°ë¦¬(km): '))
print(f'ì—°ë£ŒëŸ‰: {car1.get_fuelamount(distance)}')
```
